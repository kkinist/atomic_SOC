## Plot the dynw function (energy-dependent weighting in Molpro CASSCF)

```python
import numpy as np
import matplotlib.pyplot as plt

WAVENUM = 219474.63  # cm-1 per hartree
```

```python
dynfac = 8  # value of the dynw parameter, dynfac (hartree units)
max_cm = 50 * 1000 # maximum energy for plot, in cm-1 units
```

```python
def dynw(dynfac, E):
    # parameters are in hartree units
    x = (dynfac * E) ** 2
    w = 1 / np.cosh(x)
    return w
```

```python
e = np.linspace(0, max_cm)
w = dynw(dynfac, e / WAVENUM)
plt.plot(e, w)
plt.grid()
plt.xlabel('Relative energy / cm$^{-1}$')
plt.ylabel('Weight in CASSCF')
plt.show()
```
